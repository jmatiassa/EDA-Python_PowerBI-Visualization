## Data transforming
1. Drop duplicates
2. Date column to datetime
3. Change Price Each into Unit Price

In order to process data transforming it is going to be created a copy of the raw dataframe to work with

```python
#Copy of raw dataframe
df_transforming=df_eda_raw.copy()
```

Dropping duplicates

```python
df_transforming.drop_duplicates()
```

Date column to datetime

```python
ep.column_to_datetime(df_transforming,"Order Date")
```

Transfrorm column names

```python
ep.change_column_name(df_transforming,"Price Each","Unit Price")
```

Separate column: Purchase Adress

```python
#Regular expression formula to separate the text
df_transforming=ep.separate_address(df_transforming,"Purchase Address")
```

```python
df_transforming.head()
```

```python
#Validation of existing City column vs Address_City column
# Clean the columns to remove leading/trailing spaces and make the comparison case-insensitive
df_transforming["City"] = df_transforming["City"].str.strip().str.lower()
df_transforming["Address_City"] = df_transforming["Address_City"].str.strip().str.lower()

# Now filter where the values in 'City' and 'Address_City' are equal
df_citivalidation = df_transforming[df_transforming["City"] != df_transforming["Address_City"]]

# Display the result
print(df_citivalidation)
```

```python
# Dropping City column due to duplicity with Address_City
df_transforming=df_transforming.drop(columns='City')
```
